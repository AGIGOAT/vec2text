```python
from tqdm.notebook import trange
import torch

trainer.args.per_device_eval_batch_size = 32

d = trainer.get_eval_dataloader()
batch = next(iter(d))

# torch._dynamo.config.verbose = True
trainer.model.eval()

num_beams = 1
gen_kwargs = {
    "early_stopping": True,
    # "no_repeat_ngram_size": 3,
    # From CtRL paper: We find that using a greedy sampling 
    # and θ ≈ 1.2 yields a good balance between truthful generation
    # and lack of repetition (arxiv.org/abs/1909.05858)
    "repetition_penalty": 1.2,
    "num_beams": num_beams,
    'do_sample': False,
}

torch.set_grad_enabled(False)

input_embs = trainer.model.call_embedding_model(
    input_ids=batch['embedder_input_ids'].cuda(),
    attention_mask=batch['embedder_attention_mask'].cuda(),
)

all_data = []

trainer.model.embedder_decode_score = False
outputs = trainer.model.generate(
    inputs={k: v.cuda() for k,v in batch.items()},
    generation_kwargs=gen_kwargs
)[:, 1:]

trainer.model.embedder_decode_score = True
trainer.model.embedding_decode_score_lambda_val = 0.5 # 0.35
trainer.model.embedding_decode_score_topk = 32
outputs_reranked = trainer.model.generate(
    inputs={k: v.cuda() for k,v in batch.items()},
    generation_kwargs=gen_kwargs
)[:, 1:]

for prefix_length in trange(1, outputs.shape[1]+1):# 1, outputs.shape[1]):
    # pred prefix
    prefixes = outputs[:, :prefix_length]
    # add eos token
    prefixes = torch.cat((prefixes, torch.tensor([[trainer.model.embedder_tokenizer.eos_token_id]], device=prefixes.device).repeat(len(prefixes), 1)), dim=1)
    prefix_embs = trainer.model.call_embedding_model(
        input_ids=prefixes,
        attention_mask=torch.ones_like(prefixes),
    )
    prefix_sims = torch.nn.CosineSimilarity(dim=1)(input_embs, prefix_embs)
    all_data.extend(
        [[j, p.cpu().tolist(), sim.cpu().item(), prefix_length, "predicted"] for j, (p, sim) in enumerate(zip(prefixes, prefix_sims))]
    )
    # pred prefix w/ reranking
    prefixes_reranked = outputs_reranked[:, :prefix_length]
    # add eos token
    prefixes_reranked = torch.cat((prefixes_reranked, torch.tensor([[trainer.model.embedder_tokenizer.eos_token_id]], device=prefixes.device).repeat(len(prefixes_reranked), 1)), dim=1)
    prefix_reranked_embs = trainer.model.call_embedding_model(
        input_ids=prefixes_reranked,
        attention_mask=torch.ones_like(prefixes_reranked),
    )
    prefix_reranked_sims = torch.nn.CosineSimilarity(dim=1)(input_embs, prefix_reranked_embs)
    all_data.extend(
        [[j, p.cpu().tolist(), sim.cpu().item(), prefix_length, "predicted_reranked"] for j, (p, sim) in enumerate(zip(prefixes_reranked, prefix_reranked_sims))]
    )
    # true pre|fix
    true_prefixes = batch['embedder_input_ids'][:, :prefix_length].cuda()
    # add eos token
    true_prefixes = torch.cat((true_prefixes, torch.tensor([[trainer.model.embedder_tokenizer.eos_token_id]], device=prefixes.device).repeat(len(true_prefixes), 1)), dim=1)
    true_embs = trainer.model.call_embedding_model(
        input_ids=true_prefixes,
        attention_mask=torch.ones_like(true_prefixes),
    )
    true_sims = torch.nn.CosineSimilarity(dim=1)(input_embs, true_embs)
    all_data.extend(
        [[j, p.cpu().tolist(), sim.cpu().item(), prefix_length, "true"] for j, (p, sim) in enumerate(zip(prefixes, true_sims))]
    )
```

```python
import pandas as pd

df = pd.DataFrame(all_data, columns=["i", "prefix", "sim", "prefix_length", "ptype"])
df.head()
```
