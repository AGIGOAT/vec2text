```python
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
def get_entities(text):
    doc = nlp(text)
    named_entities = []
    for entity in doc.ents:
        named_entities.append((entity.text, entity.label_))
    return named_entities
```
