## T-test

```python
import numpy as np
from scipy import stats

method1_scores = inv_trainer.bleu_results 
method2_scores = corr_trainer.bleu_results

# Perform a t-test
t_statistic, p_value = stats.ttest_ind(method1_scores, method2_scores)

# Check the p-value to determine significance
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis: Methods have different effects.")
else:
    print("Fail to reject the null hypothesis: No significant difference between methods.")
    
print("T-statistic:", t_statistic)
print("P-value:", p_value)
```
